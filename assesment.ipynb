```python
import pandas as pd 
data = {
    "Branch": ["Branch A", "Branch B", "Branch C"],
    "Total Sales": [45000, 38500, 52000],
    "Op. Cost": [32000, 21000, 48000],
    "Cust. Feedback": [
        "Great service, but long queues.",
        "Always has stock, friendly staff.",
        "Expensive shipping, slow delivery."
    ]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['Net Profit']=df['Total Sales']-df['Op. Cost']
df['Profit Margin']=(df['Net Profit']/df['Total Sales'])*100
```

```python
df
```

```python
import plotly.express as px
fig = px.bar(
    df,
    x="Branch",
    y="Net Profit",
    title="Net Profit Comparison Across Branches",
    text="Net Profit"
)

# Improve layout
fig.update_traces(textposition='outside')
fig.update_layout(
    xaxis_title="Branches",
    yaxis_title="Net Profit ($)",
    template="plotly_white"
)

fig.show()
```

### Sentiment Analysis 


```python
def sentiment_score(text):
    positive_words = ["great", "friendly", "stock", "service"]
    negative_words = ["long", "queues", "expensive", "slow"]

    text = text.lower()
    score = 0

    for word in positive_words:
        if word in text:
            score += 1

    for word in negative_words:
        if word in text:
            score -= 1

    return score

df["Sentiment Score"] = df["Cust. Feedback"].apply(sentiment_score)

df = df.sort_values(by="Sentiment Score", ascending=False)
df["Rank"] = range(1, len(df) + 1)

print(df)
```

### Reasons for the Ranking:
- Branch B ranked the highest because the feedback contains positive phrases such as “always has stock” and “friendly staff,” indicating good customer satisfaction and service quality.
- Branch A received a medium ranking because, although customers mentioned “great service,” the issue of “long queues” reflects a negative customer experience as well.
- Branch C ranked the lowest because the feedback mainly contains negative terms such as “expensive shipping” and “slow delivery,” which indicate customer dissatisfaction.

### Manager Pitch
Based on both financial performance and customer feedback, Branch B should receive the performance bonus because it has the highest net profit margin and the most positive customer experience. Branch A is performing reasonably well financially, but improvements can be made to reduce customer wait times and long queues. Branch C needs a process improvement plan since its operating costs are very high relative to sales, and customers have reported issues with shipping costs and delivery delays
